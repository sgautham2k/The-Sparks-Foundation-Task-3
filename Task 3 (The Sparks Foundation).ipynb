## Visualization of Sales with other categories

```python
sns.boxplot(x='Category', y='Sales', data=data)
plt.yscale('log')
plt.show()
```

Observation : Here we can see that sales of products related to office supplies are on an average lower than the sales of other two categories.

## Visualisation of profits with other categories

```python
sns.boxplot(x='Category', y='Profit', data=data)
plt.show()
```

Observation : There are lot of outliers in the data due to which the boxbplots are not comprehensible. so we need to compare sales and profit using any other visualisation tool.

## Comparison of Sales and Profit with other categories of products

```python
plt.figure(figsize=(24,7))
plt.subplot(1,2,1)
sns.barplot(x='Category', y='Sales', data=data, estimator=np.median)
plt.title("Sales")
plt.subplot(1,2,2)
sns.barplot(x='Category', y='Profit', data=data, estimator=np.median)
plt.title("Profit")
plt.show()
```

Observation : So, we can see that although the sales of furnitures are high but they incurr least profit compared to office supplies and technology categories.

## Evaluation of profits incurred with respect to Shipping modes and categories

```python
plt.figure(num=None, figsize=(15,6), dpi=80, facecolor='w', edgecolor='k')
sns.barplot(x='Ship Mode', y='Profit', hue='Category', data=data, estimator=np.median)
plt.show()
```


Observation : So, we can see that for all types of shipping modes also, prodcuts related to "Technology" gets more profits.

## Profitable shipping mode

```python
sns.barplot(x='Ship Mode', y='Profit', data=data, estimator=np.median)
plt.show()
```

Observation : Thus, second class ship mode is little better than the other modes in terms of profit.

## Evaluation of profits incurred with respect to Segments and categories

```python
plt.figure(num=None, figsize=(15,6), dpi=80, facecolor='w', edgecolor='k')
sns.barplot(x='Segment', y='Profit', hue='Category', data=data, estimator=np.median)
plt.show()
```

Observation : Profit of technology products is maximum for Home office.

## Analysis of sales of sub-categories products 

```python
plt.figure(figsize=(20,10))
sns.countplot(y="Sub-Category", data=data)
plt.show()
```

Observation : We can see that Binders have maximum sales and Copiers have minimum sales.

## Visualizing sales across regions

```python
plt.figure(num=None, figsize=(15,7), dpi=65, facecolor='w', edgecolor='k')
sns.barplot(x='Region', y='Sales', data=data, estimator=np.median)
plt.show()
```
